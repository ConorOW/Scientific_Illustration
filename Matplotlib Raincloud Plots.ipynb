# Notebook for Performing Scatterplots in Matplotlib

### Notebook Setup

```python
# Import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
import ptitprince as pt
```

### Data Curation

```python
# Load in our dataset
# df = pd.read_csv('./full_dataframe_raw.csv', header=0, index_col='SubjID')
df = pd.read_csv('../USC/ENIGMA-PD/TBSS/full_dataframe.csv', header=0, index_col='SubjID')

# Drop rows where age, sex or diagnosis information is missing
df = df.dropna(subset=['diagnosis', 'sex', 'age'])

# Change column names to conform with NeuroHarmonize's api
df.rename(columns = {'site':'SITE'}, inplace = True)

# Swap out the 0s and 1s in sex
df['sex'] = df['sex'].replace({0: 'Female', 1: 'Male'})

# Swap out the 0s and 1s in diagnosis
df['diagnosis'] = df['diagnosis'].replace({0: 'Control', 1: 'PD'})

# Round ages and disease duration to two decimal places
df['age'] = df['age'].round(decimals=2)
df['disease_duration'] = df['disease_duration'].round(decimals=2)
```

### Take a look at the data

```python
# Get some information on our dataframe
subjects, variables = df.shape

print(f"Number of subjects: {subjects}")
print(f"Number of variables: {variables}")
```

### Average FA at each ENIGMA-PD Site

```python
plt.figure(figsize=(12,12))

plt.title(label="Average FA Across ENIGMA-PD Sites", fontsize=20)

sns.stripplot(y="Average_FA", 
              x="SITE", 
              data=df,
              linewidth=2,
              size=10)

pt.half_violinplot(x = "SITE",
                   y = "Average_FA",
                   data = df,
                   width = .9,
                   gridsize = 300)

plt.xlabel(xlabel="Site", fontsize=18)
plt.ylabel(ylabel="Average FA", fontsize=18)
plt.xticks(rotation=80, fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```

### Average MD at each ENIGMA-PD Site

```python
plt.figure(figsize=(12,12))

plt.title(label="Average MD Across ENIGMA-PD Sites", fontsize=20)

sns.stripplot(y="Average_MD", 
              x="SITE", 
              data=df,
              linewidth=2,
              size=10)

pt.half_violinplot(x = "SITE",
                   y = "Average_MD",
                   data = df,
                   width = .9,
                   gridsize = 300)

plt.xlabel(xlabel="Site", fontsize=18)
plt.ylabel(ylabel="Average MD", fontsize=18)
plt.xticks(rotation=75, fontsize=14)
plt.yticks(fontsize=14)

plt.show()
```
